```python
#set_keras_backend("tensorflow")
plots_dir = '%s/plots/' % (root)
results_dir = '%s/results/' % (root)
model_type = 'hpc'
station_id = 2 #param_dict['sid']
batch_size = 64 #param_dict['batch_size']
hidden_size = 1 #param_dict['nhidden']
nunits = 16 #param_dict['nunits']
dropout = 0.0 #param_dict['dropout']
start_id = 1 #param_dict['start_id'] - 1
tag = '{}-{:03d}'.format(model_type, station_id)
epochs = 2
```

```python
inp_df = pd.read_csv(root+'/data/1980-2005_2d_%03d.txt'%station_id, sep="  ", header=None, engine='python')
out_df1 = pd.read_csv(root+'/data/1980-2005_3d_vy_%03d.txt'%station_id, sep=r"\s*", header=None, engine='python')
out_df2 = pd.read_csv(root+'/data/1980-2005_3d_ux_%03d.txt'%station_id, sep=r"\s*", header=None, engine='python')
out_df3 = pd.read_csv(root+'/data/1980-2005_3d_wz_%03d.txt'%station_id, sep=r"\s*", header=None, engine='python')
out_df4 = pd.read_csv(root+'/data/1980-2005_3d_tk_%03d.txt'%station_id, sep=r"\s*", header=None, engine='python')
out_df5 = pd.read_csv(root+'/data/1980-2005_3d_qv_%03d.txt'%station_id, sep=r"\s*", header=None, engine='python')
out_df = pd.concat([out_df1, out_df2, out_df3, out_df4, out_df5], axis=1)

print(inp_df.shape)
print(out_df.shape)
print(out_df1.shape[1])

rlevels = 17 # we need 0 to 16 layers
nlevels = out_df1.shape[1]
nseries = int(out_df.shape[1]/nlevels)
tlevels = out_df.shape[1]
indices = range(tlevels)

res = []
for i in range(rlevels):
    for j in range(nseries):
        res.append(indices[i+(nlevels*j)])  
print(res)
print(len(res))
req_outputs = res
req_outputs = [int(x) for x in req_outputs]
print(req_outputs)

# load dataset
dataset = out_df
print(dataset.shape)
values = dataset.values
print(values.shape)

out_df_sel = out_df.iloc[:,req_outputs]
n_input = inp_df.shape[1]
n_output = out_df_sel.shape[1]

dataset = pd.concat([inp_df, out_df_sel], axis=1)
print(dataset.shape)
values = dataset.values
# ensure all data is float
req_data = values.astype('float32')
```

```python
preprocessor = Pipeline([('stdscaler', StandardScaler()), ('minmax', MinMaxScaler(feature_range=(0, 1)))])
preproc_data = preprocessor.fit_transform(req_data)
```

```python
# split into train and test sets
values = preproc_data
test_idx = range(40863, 72969)
train_idx = range(start_id, 40863)
val_idx = range(37946, 40863)

train = values[train_idx, :]
test = values[test_idx, :]
validation = values[val_idx, :]    

# split into input and outputs
train_X, train_y = train[:, range(n_input)], train[:, range(n_input, n_input+n_output)]
validation_X, validation_y = validation[:, range(n_input)], validation[:, range(n_input, n_input+n_output)]
test_X, test_y = test[:, range(n_input)], test[:, range(n_input, n_input+n_output)]
print(train_X.shape, train_y.shape, validation_X.shape, validation_X.shape, test_X.shape, test_y.shape)
ntrain = train_X.shape[0]
ntest = test_X.shape[0]
nval = validation_X.shape[0]
tag = '{}-{:05d}'.format(tag, ntrain)
```

```python
rlevels = 5
filepath= results_dir+tag+'_weights.best.hdf5'
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='auto')
early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='auto')
nseries_cntr = 1 
input_shape = train_X.shape[1:]
inputs = Input(shape=input_shape)
x = Dense(nunits, activation='relu')(inputs)
x = Dropout(dropout)(x)
for j in range(hidden_size):
    x = Dense(nunits, activation='relu')(x)
    x = Dropout(dropout)(x)
x = Dense(nunits, activation='relu')(x)
level_0 = Dense(nseries, name='level_0')(x)
prev = level_0
outputs_list = []
outputs_list.append(prev) 
model = Model(inputs=inputs, outputs=outputs_list)
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
model.summary()
callbacks_list = [early_stopping]
start_index = (nseries_cntr-1)*nseries
end_index = nseries_cntr*nseries
print(range(start_index, end_index))
history = model.fit(train_X, train_y[:,range(start_index, end_index)], validation_data=(validation_X, validation_y[:,range(start_index, end_index)]), callbacks=callbacks_list, epochs=epochs, batch_size=batch_size, verbose=1)
for layer in model.layers:
    layer.trainable = False

    
for i in range(rlevels-1):
    nseries_cntr = nseries_cntr + 1
    merged = concatenate([inputs] + [prev])
    x = Dense(nunits, activation='relu')(merged)
    x = Dropout(dropout)(x)
    for j in range(hidden_size):
        x = Dense(nunits, activation='relu')(x)
        x = Dropout(dropout)(x)
    x = Dense(nunits, activation='relu')(x)
    prev = Dense(nseries, name='level_%d' %(i+1))(x)
    outputs_list.append(prev)
    model = Model(inputs=inputs, outputs=prev)
    print(len(model.layers))
    #model = Model(inputs=inputs, outputs=concatenate(outputs_list))
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    model.summary()
    callbacks_list = [early_stopping]
    start_index = (nseries_cntr-1)*nseries
    end_index = nseries_cntr*nseries
    print(range(start_index, end_index))
    history = model.fit(train_X, train_y[:,range(start_index, end_index)], validation_data=(validation_X, validation_y[:,range(start_index, end_index)]), callbacks=callbacks_list, epochs=epochs, batch_size=batch_size, verbose=1)
    for layer in model.layers:
        layer.trainable = False
```
