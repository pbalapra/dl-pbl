```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
import glob
from matplotlib import pyplot

rlevels =  17

def compute_metric(y_true, y_pred):
    r2 = r2_score(y_true, y_pred) 
    evs = explained_variance_score(y_true, y_pred) 
    mae = mean_absolute_error(y_true, y_pred)
    return r2, evs, mae

def compute_tuning_metric(obsr, preds):
    result = []
    for level in range(1, rlevels):
        start = (5*level) - 5
        end = (5*level)
        groups = range(start,end)
        for group in reversed(groups):
            r2, evs, mae = compute_metric(obsr[:, group], preds[:, group])
            #print('%d %d %f %f %f' %(level, group % 5, r2, evs, mae))
            result.append([level, group % 5, r2, evs, mae])
    res = pd.DataFrame(result)
    res.columns = ['level', 'group', 'r2', 'evs', 'mae']
    return res

def compute_data_frame(obsr_filename, pred_filename):
    pred_df = pd.read_csv(pred_filename, sep=",")
    obsr_df = pd.read_csv(obsr_filename, sep=",")
    #print(pred_df.columns.values)
    ncols = pred_df.shape[1]
    pred_unique = pred_df.iloc[:,99:ncols].drop_duplicates()
    #print(pred_unique)
    preds = pred_df.iloc[:,16:101]
    obsr = obsr_df.iloc[:,16:101]
    #print(preds.shape)
    #print(obsr.shape)
    res = compute_tuning_metric(obsr.values, preds.values)
    for name in pred_unique.columns:
        res[name] = pd.DataFrame({'name': pred_unique[name].values.tolist() * res.shape[0]}) 
    return(res)
```

```python
df = pd.DataFrame()
for filename in glob.iglob('../results/*_predicted.csv', recursive=True):
    pred_filename = filename
    obsr_filename = filename.replace('predicted', 'observed')
    print(obsr_filename)
    print(pred_filename)
    df2 = compute_data_frame(obsr_filename, pred_filename)
    df1 = df.copy()
    df = df1.append(df2)
```

```python
print(df.shape)
print(df.head())
```

```python
def getDF(df, level, group, model_type, station_id, ntrain):
    df1 = df[(df['level']==level) & (df['group']==group) & (df['model_type']==model_type) & (df['ntrain']==ntrain) & (df['station_id']==station_id) ]
    return df1['r2'].values.tolist()
```

```python
getDF(df, level=16, group=4, model_type='mlp', station_id=2, ntrain=29182)
```

```python
groups = [0, 1, 2, 3, 4]
gnames = ['vy', 'ux', 'wz', 'tk', 'qv']
model_types = ['hpc', 'hac', 'mlp']
tyears = ['1994', '1992-1994', '1990-1994', '1985-1994', '1980-1994']
ntrains = list(set(df.ntrain.values))
ntrains_index = 4
for group in groups: 
    pyplot.figure(figsize=(20, 10), dpi= 80, facecolor='w', edgecolor='k')
    for method in model_types:
        vals = [getDF(df, level=i, group=group, model_type=method, station_id=2, ntrain=ntrains[ntrains_index])[0] for i in range(1,17)]
        pyplot.plot(range(1,rlevels), vals, label = method, linestyle= ':', linewidth=15)
        #pyplot.plot(vals, linestyle= ':', linewidth=5)
    pyplot.xlabel('level', fontsize=40)
    pyplot.ylabel(r'R$^2$', fontsize=40)
    pyplot.ylim(0, 1)
    pyplot.title('output=%s; training years=%s'% (gnames[group], tyears[ntrains_index]), fontsize = 40)
    pyplot.tick_params(labelsize=30)
    pyplot.legend(fontsize=30, loc=3)
    pyplot.grid()
    pyplot.show()
```

```python
method = 'hpc'
for group in groups: 
    pyplot.figure(figsize=(20, 10), dpi= 80, facecolor='w', edgecolor='k')
    for ntrains_index in range(len(ntrains)):
        vals = [getDF(df, level=i, group=group, model_type=method, station_id=2, ntrain=ntrains[ntrains_index])[0] for i in range(1,17)]
        pyplot.plot(range(1,rlevels), vals, label = tyears[ntrains_index], linestyle= ':', linewidth=15)
    pyplot.xlabel('level', fontsize=40)
    pyplot.ylabel(r'R$^2$', fontsize=40)
    pyplot.ylim(0, 1)
    pyplot.title('output=%s; method=%s'% (gnames[group], method), fontsize = 40)
    pyplot.tick_params(labelsize=30)
    pyplot.legend(fontsize=30, loc=3)
    pyplot.grid()
    pyplot.show()
```
